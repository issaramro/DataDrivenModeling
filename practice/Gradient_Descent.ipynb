```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, .5, 2, .3, .2] 
y = [55, 37.5, 89.8, 31.8, 31] 

order = 1
eta = 0.01

X = X = np.zeros((len(x), order+1))

for i in range(order+1):
    X[:, i] = np.array(x)**i

w = np.zeros(order+1)

for i in range(500):
    grad = -2*X.T@(y - X@w)
    w = w - eta*grad

p = 0
for i in range(order+1):
    p += w[i]* np.array(x)**i

plt.plot(x, p, color = 'k')
plt.scatter(x,y, color = 'b')
plt.ylabel("Length (cm)")
plt.xlabel("Weight (kg)")
plt.show()

test_weights = [1.5, .8]
pred = 0
for i in range(order+1):
    pred += w[i]* np.array(test_weights)**i
print(pred)
```
