```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# it is computationally costy to calculate the inverse

inputs = [1, .5, 2, .3, .2] 
outputs = [55, 37.5, 89.8, 31.8, 31] 


order = 1
n = len(inputs)

X = np.zeros((n, order+1))

for i in range(order+1):
    X[:, i] = np.array(inputs)**i
    
X_new = np.transpose(X) @ X

y = np.transpose(X) @ np.array(outputs)

coef = np.linalg.solve(X_new, y)

p = 0
for i in range(order+1):
    p += coef[i]* np.array(inputs)**i

plt.plot(inputs, p, color = 'k')
plt.scatter(inputs,outputs, color = 'red')
plt.ylabel("Length (cm)")
plt.xlabel("Weight (kg)")
plt.show()
```
